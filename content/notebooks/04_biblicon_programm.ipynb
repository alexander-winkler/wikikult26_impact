# Quantitative Analyse der Vortragenden der Bibliocon 2026

Die Daten für das Präsentationssystem des Bibliocon-Programms speist sich aus Schnittstellen.

Näheres hierzu lässt sich über die Developer-Tools des Browsers herausfinden.


![](./img/devtool.png)

Der API-Endpunkt für die Liste der Vorsitzenden und Referierenden ist demnach `https://bibliocon2026.planner.documedias.systems/api/program/persons`.

```python
# Import der benötigten Module
import requests
from matplotlib import pyplot as plt
from collections import Counter
```

```python
persons_url = "https://bibliocon2026.planner.documedias.systems/api/program/persons"
```

Der Server verlangt einen *User Agent* im Request-Header:

```python
headers = {
    'User-Agent' : 'API-Workshop Bibliocon2026' # String beliebig wählbar
}
```

```python
res = requests.get(persons_url, headers = headers )
```

```python
res.json()
```

## Zahl der Vortragenden

## Geschlechterverteilung

Geschlecht ist in den Daten numerisch codiert:

| Schlüssel | Bezeichnung |
| --- | --- |
| 1 | männlich |
| 2 | weiblich |
| 3 | nichtbinär |
| None | 'keine Angabe' |

```python
genderMap = {
    1 : 'männlich',
    2 : 'weiblich',
    3 : 'nichtbinär',
    None : 'keine Angabe'
}
```

```python
c = Counter([_['gender_id'] for _ in res.json()]).most_common()
xs = [ genderMap[_[0]] for _ in c ]
ys = [ _[1] for _ in c ]
```

```python
plt.bar(xs, ys)
plt.title("Geschlechterverteilung der Sprecher:innen auf der Bibliocon 2026")
```

## Institutionen

```python
institutions = [ _['institution'] for _ in res.json() ]
```

```python
n = 10
xs, ys = zip(*reversed(Counter(institutions).most_common(n)))
xs = [ _ if _ is not None else "keine Angabe" for _ in xs ]
```

```python
plt.barh(xs, ys)
plt.title(f"Zahl der Vortragenden nach Institution (Top-{n})")
```

## Städte

```python
cities = [ _['city'] for _ in res.json() ]
n = 10
xs, ys = zip(*reversed(Counter(cities).most_common(n)))
xs = [ _ if _ is not None else "keine Angabe" for _ in xs ]
plt.barh(xs, ys)
plt.title(f"Zahl der Vortragenden nach Stadt (Top-{n})")
```

## Bundesländer

```python
cities = [ _['state'] for _ in res.json() ]
n = 10
xs, ys = zip(*reversed(Counter(cities).most_common(n)))
xs = [ _ if _ is not None else "keine Angabe" for _ in xs ]
plt.barh(xs, ys)
plt.title(f"Zahl der Vortragenden nach Bundesland (Top-{n})")
```

## Länder

```python
countries = [ _['country_code'] for _ in res.json() ]
n = 15
xs, ys = zip(*Counter(countries).most_common(n))
xs = [ _ if _ is not None else "keine Angabe" for _ in xs ]
plt.bar(xs, ys)
plt.title(f"Zahl der Vortragenden nach Länder (Top-{n})")
```
