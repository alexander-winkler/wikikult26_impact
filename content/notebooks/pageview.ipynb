Page-View-API
https://doc.wikimedia.org/generated-data-platform/aqs/analytics-api/reference/page-views.html

```python
import requests
from matplotlib import pyplot as plt
```

```python
headers = {
    'User-Agent' : 'pageview test (winkler@zib.de)'
}
res = requests.get("https://wikimedia.org/api/rest_v1/metrics/pageviews/top/commons.wikimedia.org/all-access/2023/01/01", headers = headers)
```

```python
visualisierung davon
```

```python
xs = list()
ys = list()

for _ in res.json()['items'][0]['articles'][:30]:
    xs.append(_['article'])
    ys.append(_['views'])
```

```python
plt.bar(xs,ys)
```
