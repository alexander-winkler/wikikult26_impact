# Daten aus dem Wikiversum

```python
gndid = "1011521105"
query = f'''
SELECT ?person ?personLabel ?geburtstag ?sterbetag ?img WHERE {{
  ?person wdt:P569 ?geburtstag .
  OPTIONAL {{
    ?person wdt:P570 ?sterbetag 
  }}
  ?person wdt:P227  "{gndid}".
  OPTIONAL {{
    ?person wdt:P18 ?img
            }}
  SERVICE wikibase:label {{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE],mul,en". }}
  }}
'''
```

```python
import requests
```

```python
endpoint = "https://query.wikidata.org/sparql"
user_agent = "Testabruf Workshop BiblioCon 2026"
```

```python
headers = {
    'User-Agent' : user_agent,
    'Accept' : 'application/sparql-results+json'
}

params = {
    'query' : query
}
```

```python
res = requests.get(endpoint, params = params, headers = headers)
```

```python
res.json()
```

```python
for elem in res.json()['results']['bindings']:
    wd_uri = elem['person']['value']
    geburtstag = elem['geburtstag']['value']
    try:
        sterbetag = elem['sterbetag']['value']
    except:
        sterbetag = None
    try:
        img = elem['img']['value']
    except:
        img = None
```
