## Iteration

```python
for element in meineListe:
    print(element)
```

```python
neue_liste = []
for element in meineListe:
    neue_liste.append(element + "_neu")
```

```python
print(neue_liste)
```

## Dictionaries

```python
mein_dictionary = {}

for element in meineListe:
    key = element
    value = len(element)
    mein_dictionary[key] = value
```

```python
mein_dictionary
```

```python
mein_dictionary['ist']
```

## Counter

```python
from collections import Counter
```

```python
buchstabenliste = []
for element in neue_liste:
    buchstabenliste.extend(element)

print(buchstabenliste)
```

```python
Counter(buchstabenliste)
```

## HTML-Parsing

```python
from lxml import etree
```

```python
root = etree.HTML(response.text)
```

```python
root
```

```python
for element in root.iter():
    print(element)
```

```python
# Linkliste

links = root.xpath('//a')
```

```python
links_dict = {}
for l in links:
    href = l.attrib['href']
    text = l.text
    links_dict[href] = text
```
